```python
import math
import numpy as np
import matplotlib.pyplot as plt

print("введите R")
R=int(input())

print("введите r")
r=int(input())

a=[]

b=[]

for fi in np.arange(0,30,0.01):
    X = (R-r)*math.cos(fi) + r * math.cos(fi-(r/R) * fi )   
    Y = (R-r)*math.sin(fi) + r * math.sin(fi-(r/R) * fi ) 
    
    a.append(X)        
    b.append(Y)
```

```python
import matplotlib.pyplot as plt

#plt.plot(X,Y,lw=2)
#plt.axis('equal')
#plt.show()

plt.plot(a, b)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Фигура')
plt.grid(True)
plt.show()
```
