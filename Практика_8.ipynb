```python
import numpy as np
import matplotlib.pyplot as plt

def predator_prey_model(alpha, beta, gamma, delta, x0, y0, t_end, dt):
    t = np.arange(0, t_end, dt)
    n = len(t)
    x = np.zeros(n)
    y = np.zeros(n)
    x[0] = x0
    y[0] = y0

    for i in range(1, n):
        dx_dt = alpha * x[i-1] - beta * x[i-1] * y[i-1]#dx/dt и сама производная 
        dy_dt = -gamma * y[i-1] + delta * x[i-1] * y[i-1]#dy/dt и сама производная
        x[i] = x[i-1] + dx_dt * dt#Обновляем значения популяций жертв и хищников на текущем временном шаге, 
        y[i] = y[i-1] + dy_dt * dt#добавляя изменение (dx_dt и dy_dt) умноженное на шаг времени (dt).

    return t, x, y

# Параметры модели
alpha = 1.0#коэффициент рождаемости жертв
beta = 0.5#коэффициент убийств
gamma = 0.2#коэффициент убыли хищников
delta = 0.1#коэффициент рождаемости хищников

#начальное количестиво Хищников и жертв
x0 = 10.0
y0 = 5.0

#конечное время модели и её шаг
t_end = 100.0
dt = 0.1

# Вычисление модели
t, x, y = predator_prey_model(alpha, beta, gamma, delta, x0, y0, t_end, dt)

# Построение графика
plt.plot(t, x, label='Жертвы')
plt.plot(t, y, label='Хищники')
plt.xlabel('Время')
plt.ylabel('Популяция')
plt.title('Модель хищник-жертва')
plt.legend()
plt.show()
```
