```python
import numpy as np
import matplotlib.pyplot as plt

def predator_prey_model(alpha, beta, gamma, delta, x0, y0, t_end, dt):
    t = np.arange(0, t_end, dt)
    n = len(t)
    x = np.zeros(n)
    y = np.zeros(n)
    x[0] = x0
    y[0] = y0

    for i in range(1, n):
        dx_dt = alpha * x[i-1] - beta * x[i-1] * y[i-1]#dx/dt    
        dy_dt = -gamma * y[i-1] + delta * x[i-1] * y[i-1]#dy/dt   
        x[i] = x[i-1] + dx_dt * dt#         , 
        y[i] = y[i-1] + dy_dt * dt#  (dx_dt  dy_dt)     (dt).

    return t, x, y

#  
alpha = 1.0#  
beta = 0.5# 
gamma = 0.2#  
delta = 0.1#  

#    
x0 = 10.0
y0 = 5.0

#     
t_end = 100.0
dt = 0.1

#  
t, x, y = predator_prey_model(alpha, beta, gamma, delta, x0, y0, t_end, dt)

#  
plt.plot(t, x, label='')
plt.plot(t, y, label='')
plt.xlabel('')
plt.ylabel('')
plt.title(' -')
plt.legend()
plt.show()
```
