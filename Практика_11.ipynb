–ø–æ—Å–ª–µ —á–µ–≥–æ –Ω–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ s –æ–±—ä—è–≤–ª—è–µ—Ç—Å—è –æ—Ç–≤–µ—Ç–æ–º.

32, –∞ —á–∏—Å–ª–∞ a, c –∏ –Ω–∞—á–∞–ª—å–Ω–æ–µ —Å–æ—Å—Ç–æ—è–Ω–∏–µ s –∑–∞–¥–∞–Ω—ã –≤–æ 
–≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö. 
–í—ã–≤–µ–¥–∏—Ç–µ —Å–ª–µ–¥—É—é—â–∏–µ 10 —á–∏—Å–µ–ª, –∫–æ—Ç–æ—Ä—ã–µ —Å–≥–µ–Ω–µ—Ä–∏—Ä—É–µ—Ç —ç—Ç–æ—Ç –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä.
–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω—ã —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª —Ç—Ä–∏ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞: a, c –∏ s. 
–ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ —ç—Ç–∏ —á–∏—Å–ª–∞
–ø–æ–º–µ—â–∞—é—Ç—Å—è –≤ 32-–±–∏—Ç–Ω—ã–π –±–µ–∑–∑–Ω–∞–∫–æ–≤—ã–π —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö.
–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–í—ã–≤–µ–¥–∏—Ç–µ 10 —Å–ª–µ–¥—É—é—â–∏—Ö —á–∏—Å–µ–ª, –∫–æ—Ç–æ—Ä—ã–µ —Å–≥–µ–Ω–µ—Ä–∏—Ä—É–µ—Ç –ø–æ–ª—É—á–∏–≤—à–∏–π—Å—è 
–≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä

```python
def linear_congruential_generator(a, c, s, m, n):
    result = []
    for _ in range(n):
        s = (s * a + c) % m
        result.append(s)
    return result

# –í–≤–æ–¥ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–∞ 
a, c, s = map(int, input().split())

# –ì–µ–Ω–µ—Ä–∞—Ü–∏—è —Å–ª–µ–¥—É—é—â–∏—Ö 10 —á–∏—Å–µ–ª
generated_numbers = linear_congruential_generator(a, c, s, 2**32, 10)

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
for number in generated_numbers:
    print(number)
```
