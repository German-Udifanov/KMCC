#Фигура Эпитрохойда

```python
print("Введите R")
R=int(input())

print("Введите r")
r=int(input())

print("Введите h")
h=int(input())

#my_float = 3.14
#masx = []
#masx.append(my_float)

#my_float1 = 3.14
#masy = []
#masy.append(my_float1)

masx=[]
masy=[]


for fi in np.arange(0,30+1,0.01):
    x = (R+r)*math.cos(r/R*fi)-h*math.cos((R+r)/r*fi)
    y = (R+r)*math.sin(r/R*fi)-h*math.sin((R+r)/r*fi)
    
    masx.append(x)
    masy.append(y)
```

```python
import matplotlib.pyplot as plt

plt.plot(masx, masy)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Фигура Эпитрохойда')
plt.grid(True)
plt.show()
```

фигура Гипотрохойда

```python
print("Введите R")
R=int(input())

print("Введите r")
r=int(input())

print("Введите h")
h=int(input())

masx=[]
masy=[]

for fi in np.arange(0,30+1,0.01):
    x=(R-r)*math.cos(fi)+r*math.cos(fi-(r/R)*fi)   
    y=(R-r)*math.sin(fi)+r*math.sin(fi-(r/R)*fi)
    
    masx.append(x)
    masy.append(y)
 
```

```python
plt.plot(masx, masy)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Фигура Гипотрохойда')
plt.grid(True)
plt.show()
```
