```python
import numpy as np
import math

N=np.arange(1,10+1,1)

T0=[0.001,0.001,0.001,0.001,0.001,0.01,0.01,0.01,0.01,0.01]

a=[1.0,0.9,0.8,0.7,0.6,1.0,0.9,0.8,0.7,0.6]
    
#w=np.arange(0.2,0.4+0.05,0.05)

w=[0.2, 0.25, 0.3, 0.35, 0.4, 0.2, 0.25, 0.3, 0.35, 0.4]

#fi0=np.arange(0,0.2,0.05)

fi0=[0,0.05,0.1,0.15,0.2,0,0.05,0.1,0.15,0.2]

gam=np.arange(0.01,0.1+0.01,0.01)

wzat=[]
fi=[]
tn=[]

for i in range(0,10):
    
    wzat.append(math.sqrt(w[i]**2-gam[i]**2))
    
    tn.append(T0[i]*N[i])
    
    fi.append(wzat[i]*tn[i]+fi0[i])

y=[]

for i in range(0,10):
    y.append(a[i]*(math.e**(-gam[i]*tn[i]))*math.sin(fi[i]))

print(len(y))
```
